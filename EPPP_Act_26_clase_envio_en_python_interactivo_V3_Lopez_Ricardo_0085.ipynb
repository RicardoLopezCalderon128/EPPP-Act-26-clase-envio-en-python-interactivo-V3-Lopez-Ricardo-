```python
# --- CLASE ENVIO EN PYTHON ---

class Envio:
    def __init__(self):
        # Atributos inicializados
        self.destino = ""
        self.peso = 0.0
        self.costo = 0.0

    # Función para capturar los datos (Interactiva)
    def leerdatos(self, etiqueta):
        print(f"\n--- Datos del {etiqueta} ---")
        self.destino = input("Ciudad de destino: ")
        self.peso = input("Peso del paquete (kg): ")
        self.costo = input("Costo del envío ($): ")

    # Función para mostrar la información
    def mostrardatos(self, etiqueta):
        print(f"\n📦 RESUMEN DEL {etiqueta.upper()}:")
        print(f"📍 Destino: {self.destino}")
        print(f"⚖️ Peso: {self.peso} kg")
        print(f"💰 Costo: ${self.costo}")

# --- EJECUCIÓN DEL PROGRAMA ---

print("=== SISTEMA DE REGISTRO DE ENVÍOS EN PYTHON ===")

# 1. Creación de los 2 objetos (Instanciación)
envio1 = Envio()
envio2 = Envio()

# 2. Llamada a funciones para leer datos de forma interactiva
envio1.leerdatos("Envío 1")
envio2.leerdatos("Envío 2")

# 3. Llamada a funciones para mostrar los resultados
print("\n" + "="*40)
print("     RESULTADOS DEL REGISTRO")
print("="*40)

envio1.mostrardatos("envio1")
envio2.mostrardatos("envio2")
print("programa realizado poe Ricardo Lopez NL = 29")
print("\n================================================")
```
